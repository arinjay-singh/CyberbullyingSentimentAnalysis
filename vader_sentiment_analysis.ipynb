```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
twitter_df = pd.read_csv('cyberbullying_tweets.csv')
twitter_df['neg'] = np.nan
twitter_df['neu'] = np.nan
twitter_df['pos'] = np.nan
twitter_df['compound'] = np.nan
twitter_df['cb_label'] = np.nan
```

```python
for index, row in twitter_df.iterrows():
    vs = analyzer.polarity_scores(row['tweet_text'])
    twitter_df.loc[index, 'neg'] = vs['neg']
    twitter_df.loc[index, 'neu'] = vs['neu']
    twitter_df.loc[index, 'pos'] = vs['pos']
    twitter_df.loc[index, 'compound'] = vs['compound']
    twitter_df.loc[index, 'cb_label'] = 1 if vs['compound'] < -0.5 else 0
twitter_df['cb_label'] = twitter_df['cb_label'].astype(int)
```
